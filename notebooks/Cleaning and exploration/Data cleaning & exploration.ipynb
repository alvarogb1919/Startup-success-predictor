### Funding_total_usd needs to be an integer;  Founded_at, first_funding_at, and last_funding_at should be a date, let's change those via functions

```python
def funding_conversor(x):
    if x in ["-", " -   "]:
        return 0
    else:
        split = x.split(",")
        return int(''.join(split))
```

```python
startups['funding_total_usd'] = startups.funding_total_usd.apply(funding_conversor)
```

### I am now going to add a new column that will represent the time difference between the founding date and the date the startup received it's first funding

```python
startups['delta_founded_vs_firstfunding'] = startups.first_funding_at - startups.founded_at
```

```python
startups.head()
```

```python
startups.dtypes
```

```python
import numpy as np
```

```python
def days_to_integer(x):
    return x / np.timedelta64(1, 'D')
```

```python
startups['delta_founded_vs_firstfunding'] = startups.delta_founded_vs_firstfunding.apply(days_to_integer)
```

```python
startups.head()
```

### For this new column there are some negative values which don't make sense as it's not possible to receive funding wihout existing, hence, I will adjust the dataset for those cases

```python
startups.loc[startups.delta_founded_vs_firstfunding < 0]
```

### Let's exclude those rows from the dataset

```python
startups = startups.loc[startups.delta_founded_vs_firstfunding >= 0]
```
