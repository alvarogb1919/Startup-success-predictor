### El link a dealroom es el siguiente, lo que habria que hacer es un loop para añadir "/startup name" al link y que obtenga el valuation de ahi

```python
import requests
```

```python
from bs4 import BeautifulSoup
```

```python
url_dr = "https://app.dealroom.co/companies"
```

```python
url_list_dr = []
```

```python
company_list = list(startups["name"])
```

```python
for c in company_list:
    company_url = url_dr + f"/{c}"
    url_list_dr.append(company_url)
```

```python
url_list_dr[:5]
```

```python
def obtain_valuation(u):
    response = requests.get(u)
    soup = BeautifulSoup(response.text, 'html.parser')
    company_funding = (soup.find_all(name="span", class_="valuation__value")[0].text)
    return company_funding
```

```python
obtain_valuation("https://app.dealroom.co/companies/stripe")
```
