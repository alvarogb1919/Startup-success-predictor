### Let's train another  model with the variables we have so far

```python
import pandas as pd
```

```python
pd.set_option("display.max_columns", None)
startups = pd.read_csv("../../notebooks/Web_scraping/startups_with_valuation_v_final.csv")
```

```python
startups.head()
```

```python
startups.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
startups.sample(5)
```

### My predictor is going to be a quantitative column "valuation", and the variables are going to be all the numerical ones

```python
startups.dtypes
```

### There are some categorical values that I would like to take into account, such as the country or market, for that we will use rank encoding initially

```python
top_20_countries = startups.country_code.value_counts().head(20)
```

```python
top_20_countries_df = pd.DataFrame(top_20_countries)
```

```python
top_20_countries_df["rank"] = range(1, len(top_20_countries_df)+1)
```

```python
top_20_countries_df.drop('country_code', axis=1, inplace=True)
```

```python
countries = top_20_countries_df.to_dict(orient="index")
```

```python
countries["USA"]['rank']
```

```python
top_20_countries_list = countries.keys()
```

```python
def country_ranking(x):
    if x in top_20_countries_list:
        return countries[x]['rank']
    else:
        return 21
```

```python
ranking_country = []
```

```python
for i, row in startups.iterrows():
    name = row["country_code"]
    ranking = country_ranking(name)
    ranking_country.append(ranking)
    
```

```python
ranking_country[:20]
```

```python
startups["country_ranking"] = ranking_country
```

```python
startups.head()
```

### I have elaborated a rank encoding column for the country of each startup based on the frequency of the country within the dataset. Now I will perform the same operation but for the market

```python
top_20_markets = startups.market.value_counts().head(20)
```

```python
top_20_markets
```

```python
top_20_markets_df = pd.DataFrame(top_20_markets)
```

```python
top_20_markets_df["rank"] = range(1, len(top_20_markets_df)+1)
```

```python
top_20_markets_df.drop('market', axis=1, inplace=True)
```

```python
markets = top_20_markets_df.to_dict(orient="index")
```

```python
markets[" Finance "]["rank"]
```

```python
top_20_markets_list = markets.keys()
```

```python
def market_ranking(x):
    if x in top_20_markets_list:
        return markets[x]['rank']
    else:
        return 21
```

```python
ranking_market = []
```

```python
for i, row in startups.iterrows():
    name = row["market"]
    ranking = market_ranking(name)
    ranking_market.append(ranking)
```

```python
ranking_market[:20]
```

```python
startups["market_ranking"] = ranking_market
```

```python
startups.head()
```

```python
startups = startups[startups["valuation"].notnull()]
```

### Let's now drop those columns that we will not be using

```python
startups.drop(['permalink', 'name', 'market','status', 'country_code', 'city', 'founded_at', 'first_funding_at', 'last_funding_at'], axis=1, inplace=True)
```

```python
X = startups.drop('valuation', axis=1)
```

```python
y = startups['valuation']
```

```python
from sklearn.model_selection import train_test_split as tts
```

```python
X_train, X_test, y_train, y_test = tts(X, y, test_size=0.2)
```

### Let's train a Simple Linear Regression

```python
from sklearn.linear_model import LinearRegression as LinReg
```

```python
linreg=LinReg()
```

```python
linreg.fit(X_train, y_train) 
```

```python
linreg.coef_
```

```python
linreg.intercept_
```

```python
y_pred=linreg.predict(X_test)
```

```python
y_pred[:10]
```

```python
y_test[:10]
```

### Let's take a look at the metrics

```python
from sklearn import metrics
import numpy as np
```

```python
'MAE - Error Medio Absoluto', metrics.mean_absolute_error(y_test, y_pred)
```

```python
'MSE - Error Cuadratico Medio', metrics.mean_squared_error(y_test, y_pred)
```

```python
'RMSE - Raiz Error Cuadratico Medio', metrics.mean_squared_error(y_test, y_pred)**0.5
```

```python
'R2 - Coeficiente de Determinacion', metrics.r2_score(y_test, y_pred)
```

### This model doesn't look very promising at all
