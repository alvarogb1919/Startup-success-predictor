### Let's train an initial simple model with the variables we have so far

```python
import pandas as pd
```

```python
pd.set_option("display.max_columns", None)
startups = pd.read_csv("../../dataset/Cleaned_startups.csv")
```

```python
startups.head()
```

```python
startups.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
startups.sample(5)
```

### My predictor is going to be a multiclass column "status", and the variables are going to be all the numerical ones

```python
startups.dtypes
```

```python
startups.drop(['permalink', 'name', 'market', 'country_code', 'city', 'founded_at', 'first_funding_at', 'last_funding_at'], axis=1, inplace=True)
```

```python
X = startups.drop('status', axis=1)
```

```python
y = startups['status']
```

```python
from sklearn.model_selection import train_test_split as tts
```

```python
X_train, X_test, y_train, y_test = tts(X, y, test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_1 = LogisticRegression()
```

```python
log_1.fit(X_train, y_train)
```

```python
y_pred_test = log_1.predict(X_test)
```

```python
y_pred_train = log_1.predict(X_train)
```

### Let's take a look at some metrics now

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, fbeta_score
```

```python
accuracy_test = accuracy_score(y_true = y_test, y_pred=y_pred_test)
```

```python
accuracy_train = accuracy_score(y_true = y_train, y_pred=y_pred_train)
```

```python
accuracy_test.round(3)
```

```python
accuracy_train.round(3)
```
